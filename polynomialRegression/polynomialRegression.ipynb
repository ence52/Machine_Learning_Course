```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("salaries.csv")
```

```python
x=data.iloc[:,1:2]
y=data.iloc[:,2:]
X=x.values
Y=y.values
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

LinearRegression test

```python
plt.scatter(x,y)
plt.plot(x,lin_reg.predict(x),color="red")
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
```

```python
x_poly=poly_reg.fit_transform(X)
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
plt.scatter(X,Y)
plt.plot(X,lin_reg2.predict(poly_reg.fit_transform(X)),color="red")
```

```python
print(lin_reg2.predict(poly_reg.fit_transform([[8]])))
print(lin_reg2.predict(poly_reg.fit_transform([[5]])))
print(lin_reg2.predict(poly_reg.fit_transform([[3.1]])))
```
